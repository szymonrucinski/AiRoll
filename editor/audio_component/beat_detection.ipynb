```python
print('x length = {}'.format(len(x)))
print('sample rate = {}'.format(sr))
print('sound clip is {} seconds long' .format(int((len(x)/sr))))
t = np.array(range(0, len(x))) /sr
```

```python
indexes = detect_peaks(x, mph = 0.9, mpd=sr/2)
print('There are {} peaks'.format((len(indexes))))
```

```python
#audio from peaks
sample = np.zeros(len(x))
for index in indexes:
    sample[index] = 1

m = np.max(np.abs(sample))
sigf32 = (sample/m).astype(np.float32)
write("D:\Programowanie\AI\editor\data\samples\outputs\detect_peaks.wav", sr , sigf32)
plt.plot(t,sigf32)
```
