```python
path = Path("D:\Programowanie\AI\shot_classifier")
```

```python
# item_tfms=Resize(128, ResizeMethod.Squish)
tfms = aug_transforms(do_flip = False, flip_vert = False, mult=0)
data= ImageDataLoaders.from_folder(path,train = "train", valid_pct=0.8, item_tfms=Resize(1024), batch_tfms=tfms, bs = 30, num_workers = 4)
```
